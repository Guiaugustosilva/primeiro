```python
UPP.columns = UPP.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
UPP.columns
```

```python
UPP['mes'] = UPP['mes'].astype(int) 
```

```python
months_map = {1: 'JAN', 2: 'FEV', 3: 'MAR', 4: 'ABR', 5: 'MAI', 6:'JUN', 7:'JUL', 8: 'AGO', 9: 'SET', 10: 'OUT', 11: 'NOV', 12: 'DEZ'}
import calendar

d = dict(enumerate(calendar.month_abbr))
UPP['mes'] = UPP['mes'].map(d)
UPP
```

```python
UPP = UPP.groupby(['upp', 'ano', 'mes']).sum()
UPP
```
